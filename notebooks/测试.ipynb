```python
import sqlite3
from pathlib import Path

# 明确指定数据库路径
db_path = Path(r"E:\lianghuagit\quant-a-share\quant-a-share\data_cache\market_data.sqlite")
print("DB_PATH =", db_path)

conn = sqlite3.connect(db_path)
cur = conn.cursor()

# 列出现有的所有表
cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cur.fetchall()
print("Tables:", tables)

# 如果 stock_daily 存在，再看字段结构
if ('stock_daily',) in tables:
    print("\n=== stock_daily schema ===")
    cur.execute("PRAGMA table_info(stock_daily)")
    print(cur.fetchall())

conn.close()
```
